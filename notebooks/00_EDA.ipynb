```python
def summary_categoricals(df):
    summary = []

    cat_cols = df.select_dtypes(include=["object", "category"]).columns

    for col in cat_cols:
        total = len(df)
        nunique = df[col].nunique(dropna=True)
        top_val = df[col].value_counts(dropna=False).head(1)
        top_value = top_val.index[0]
        top_pct = top_val.iloc[0] / total * 100
        unknown_count = (df[col] == "Unknown").sum()
        unknown_pct = unknown_count / total * 100
        nan_pct = df[col].isna().sum() / total * 100

        summary.append({
            "column": col,
            "dtype": df[col].dtype,
            "% unique": f"{(nunique/total)*100:.1f}%",
            "top value": top_value,
            "% top": f"{top_pct:.1f}%",
            "% Unknown": f"{unknown_pct:.1f}%",
            "% NaN": f"{nan_pct:.1f}%"
        })

    return pd.DataFrame(summary).sort_values("% Unknown", ascending=False)

summary_categoricals(df)
```

Columnas como profesion, area, tipo_actifisica, dominancia, contextura tienen más del 50% "Unknown".
No son basura, pero aportan poco en análisis general.

```python
def summary_numerics(df):
    summary = []

    num_cols = df.select_dtypes(include=["number"]).columns

    for col in num_cols:
        total = len(df)
        nan_pct = df[col].isna().sum() / total * 100

        summary.append({
            "column": col,
            "dtype": df[col].dtype,
            "min": df[col].min(),
            "max": df[col].max(),
            "mean": df[col].mean(),
            "median": df[col].median(),
            "% NaN": f"{nan_pct:.1f}%"
        })

    return pd.DataFrame(summary).sort_values("% NaN", ascending=False)

summary_numerics(df)
```

```python
def show_unique_values(df):
    cat_columns = df.select_dtypes(include=["object", "category"]).columns

    for col in cat_columns:
        print(f"\n{col} ({df[col].nunique(dropna=True)} únicos):")
        unique_vals = df[col].dropna().unique()
        print(sorted([str(val) for val in unique_vals]))
```
